```python
import pickle
import numpy as np
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
from google.colab import drive
drive.mount('/content/drive')
caminho = "/content/drive/MyDrive/MaterialApoio/"
```

```python
with open(caminho+"credit.pkl", 'rb') as f:
  X_credit_treinamento, Y_credit_treinamento, X_credit_teste, Y_credit_teste = pickle.load(f)
```

```python
X_credit = np.concatenate((X_credit_treinamento, X_credit_teste), axis = 0)
X_credit.shape
```

```python
Y_credit = np.concatenate((Y_credit_treinamento, Y_credit_teste), axis = 0)
Y_credit.shape
```

```python
param = {'criterion':['gini','entropy'],
         'splitter':['best', 'random'],
         'min_samples_split': [2, 5, 10],
         'min_samples_leaf': [1, 5, 10]}
```

```python
gri_search = GridSearchCV(estimator = DecisionTreeClassifier(),
                          param_grid = param)
gri_search.fit(X_credit, Y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

```python
parameters = {'criterion':['gini','entropy'],
         'n_estimators':[10, 40, 100, 150],
         'min_samples_split': [2, 5, 10],
         'min_samples_leaf': [1, 5, 10]}
```

```python
gri_search = GridSearchCV(estimator = RandomForestClassifier(),
                          param_grid = parameters)
gri_search.fit(X_credit, Y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

```python
parameters = {'tol':[0.001, 0.0001, 0.00001],
         'C':[1.0, 1.5, 2.0],
         'kernel': ['rbf', 'linear', 'poly', 'sigmoid']}
```

```python
gri_search = GridSearchCV(estimator = SVC(),
                          param_grid = parameters)
gri_search.fit(X_credit, Y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

```python
parameters = {'activation':['relu', 'logistic', 'tahn'],
         'solver':['adam', 'sgd'],
          'batch_size':[10, 56]}
```

```python
gri_search = GridSearchCV(estimator = MLPClassifier(),
                          param_grid = parameters)
gri_search.fit(X_credit, Y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

```python
resultados_arvore = []
resultados_random_forest = []
resultados_knn = []
resultados_logistica = []
resultados_svm = []
resultados_rede_neural = []

for i in range(30):
  print(i)
  kfold = KFold(n_splits = 10, shuffle = True, random_state = i)

  arvore = DecisionTreeClassifier(criterion = 'entropy', min_samples_leaf = 1, min_samples_split=5, splitter='best')
  scores = cross_val_score(arvore, X_credit, Y_credit, cv = kfold)
  resultados_arvore.append(scores.mean())
```
