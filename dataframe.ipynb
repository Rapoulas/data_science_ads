```python
datas = {'nome': ['Argentina', 'Brasil', 'França', 'Italia', 'Reino Unido'],
         'continente': ['America', 'America', 'Europa', 'Europa', 'Europa'],
         'extensao': [2780, 8511, 644, 301, 244],
         'corVerde': [0, 1, 0, 1, 0]
         }

codes = ['AR', 'BR', 'FR', 'IT', 'UK']

countries = pd.DataFrame(datas, index=codes)

#Testa se um dado rótulo de linha existe
has_BR = 'BR' in countries.index
has_US = 'US' in countries.index
print("Existe o rótulo 'BR'? -> ", has_BR)
print("Existe o rótulo 'US'? -> ", has_US)
print("----------------------------------")
#Testa se um dado rótulo de coluna existe
has_greenColor = 'corVerde' in countries.columns
has_blueColor = 'corAzul' in countries.columns
print("existe a coluna 'corVerde'? -> ", has_greenColor)
print("existe a coluna 'corAzul'? -> ", has_blueColor)
print("----------------------------------")
#testa se o valor faz parte de uma coluna
has_Brasil = countries['nome'].isin(['Brasil'])
print("Existe o valor 'Brasil' na coluna 'nome'?")
print(has_Brasil)
```

```python
# Modificando o Dataframe

print("--- Dataframe original ---")
print(countries)

# inserindo o país japão
countries.loc['JP'] = {'nome': 'Japão', 'continente': 'Ásia', 'extensao': 372, 'corVerde': 0}

print('--- Dataframe com o japão ---')
print(countries)

#Altera a extensão do brasil
countries.at['BR', 'extensao'] = 8512
print('--- Dataframe com extensão brasil modificado ---')
print(countries)

# Remove Argentine e Reino Unido
countries = countries.drop(['AR', 'UK'])
print('--- Dataframe sem argentina/reino unido')
print(countries)
```
